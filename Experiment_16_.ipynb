```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #Used as a Advanced import
```

```python
# Dataset for Perforrmance of a Student
data={
    'Days':['Mon','Tue','Wed','Thrs','Fri'],
    'Study_Hrs':[3,4,7,9,8],
    'Marks':[50,60,79,90,70],
    'Attendance':[75,75,60,80,95],
    'Sleep_Hrs':[6,7,5,6,7],
    'Assignments_Completed':[1,2,2,1,3]
}
df=pd.DataFrame(data)
df
```

```python
# Line chart
plt.plot(df['Days'],df['Study_Hrs'],marker='o')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study Hrs Trend')
plt.show()
```

```python
# Advanced Line Chart
plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hrs'],marker='o',color='blue',label='Study Hrs')
plt.plot(df['Days'],df['Marks'],marker='o',color='red',label='Marks')
plt.title('Study Hrs and Marks Trend')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#Bar Chart(Comparison)
plt.bar(df['Days'],df['Marks'],color='black',label='Marks')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.legend()
plt.show()
```

```python
# Advanced Bar Graph
bars=plt.bar(df['Days'],df['Marks'],color='green',label='Marks')
for bar in bars:
    y=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,y,str(y),ha='center',va='bottom')

plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.legend()
plt.grid(axis='y')# Adds horizontal grid for readablity
plt.show()
```

```python
#3 Histogram(Distribution)
data1={
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,82,88]
}
df=pd.DataFrame(data1)
plt.hist(df['Marks'],bins=3,color='grey')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
#4 Scatter Plot(Relationship)
plt.scatter(df['Study_Hrs'],df['Marks'],color='red')
plt.title('Study_Hrs vs Marks')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.legend()
plt.show()
```

```python
# Conditional Scatter Plot
# Example Category
df['Result']=['Pass','Fail','Pass','Pass','Fail']
colors=['blue' if result=='Pass' else 'red' for result in df['Result']]
plt.scatter(df['Study_Hrs'],df['Marks'],c=colors)
plt.title('Study_Hrs vs Marks (Resultwise)')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

## Same Graphs using Seaborn (Better Visualization)

```python
# Line Plot
sns.lineplot(x='Days',y='Study_Hrs',data=df)
plt.title('Study Hrs Trend')
plt.show()
```

```python
# Histogram
sns.histplot(df['Marks'],bins=4,color='grey')
plt.title('Seaborn Histogram')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(x='Study_Hrs',y='Marks',data=df)
plt.title('Seaborn Scatter Plot')
```
